```python
import numpy as np
import pandas as pd

input_file = "kc_house_data.csv"
df = pd.read_csv(input_file)
```

```python
df.info
```

```python
housing_prepared = df.drop(["price", "date"], axis = 1)
housing_prepared["sqft_living_bedrooms"] = housing_prepared["sqft_living"] * housing_prepared["bedrooms"]
print(housing_prepared.head())
housing_labels = df["price"].copy()
```

```python
def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_scores = cross_val_score(lin_reg, housing_prepared, housing_labels, scoring = "neg_mean_squared_error", cv=10)

lin_rmse_scores = np.sqrt(-lin_scores)
display_scores(lin_rmse_scores)
```
